### Diferencia entre Borde y Contorno
La diferencia entre un borde y un contorno en OpenCV es que un contorno son todos los puntos continuos con la misma intensidad en una imagen y son bastante útiles para encontrar figuras entro de una imagen mientras que los bordes representan los cambios en las superficies entre un pixel y otro, esto se calcula por medio del gradiente entre 2 pixeles adyacentes. 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Imagen Original

```python
input_image = cv2.imread("./imagenes/imagen1.jpg", cv2.IMREAD_COLOR)
input_image_gray = cv2.cvtColor(input_image,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
plt.show()
```

### Detección de bordes con el filtro de Canny
Al aplicar la detección de bordes sobre la imagen de las figuras geometricas, podemos ver que los bordes son idetnficiados como la linea interna y externa que forma las figuras geometricas ya que en esta es donde se detectan los cambios de intensidad entre pixeles por lo que se dibujan lineas al rededor del contorno.

```python
plt.figure(figsize=(8,8))
plt.imshow(cv2.Canny(input_image_gray,100,200),cmap='gray')
plt.show()
```

### Detección de contornos
Para la detección de contornos, es recomendado binarizar la imagen para tener mejores resultados. Como podemos ver en la siguiente imagen, los contornos son pintados sobre las lineas que formas las figuras geométricas, a diferencia de los bordes que se identificaron como 2 líneas al rededor del contorno.

```python
plt.figure(figsize=(8,8))
_,thresh = cv2.threshold(input_image_gray,127,255,0)
contours, hierarchy  = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(input_image, contours, -1, (0,255,0), 3)
plt.imshow(input_image)
plt.show()
```
