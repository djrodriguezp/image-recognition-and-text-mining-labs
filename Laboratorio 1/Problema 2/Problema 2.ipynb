```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Se define la funci√≥n para unir las imagenes.

```python
def create_color_image(input_image_blue, input_image_red, input_image_green):
    width = input_image_blue.shape[0]
    height = input_image_blue.shape[1]
    output_photo = np.zeros((width, height, 3))

    for j in range(0,width):
        for i in range(0,height):
            blue = input_image_blue[j,i]
            green = input_image_green[j,i]
            red = input_image_red[j,i]

            output_photo[j,i] = [blue,green,red]

    return output_photo
```

### Escala de grises canal azul

```python
input_image_blue = cv2.imread("./imagen2/imagen2_salida_gray_azul.jpg", cv2.COLOR_BGR2GRAY)
plt.imshow(input_image_blue, cmap='gray')
```

### Escala de grises canal rojo

```python
input_image_red = cv2.imread("./imagen2/imagen2_salida_gray_rojo.jpg", cv2.COLOR_BGR2GRAY)
plt.imshow(input_image_red, cmap='gray')
```

### Escala de grises canal verde

```python
input_image_green = cv2.imread("./imagen2/imagen2_salida_gray_verde.jpg", cv2.COLOR_BGR2GRAY)
plt.imshow(input_image_green, cmap='gray')
```

### Imagen con ganales unificados

```python
output_file = "./output.jpg"

cv2.imwrite(output_file, create_color_image(input_image_blue,input_image_red,input_image_green) )
output_image = cv2.imread(output_file, cv2.IMREAD_COLOR)
output_image = cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB)
plt.imshow(output_image)
```
