```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def image_to_grayscales(input_image):
    width = input_image.shape[0]
    height = input_image.shape[1]
    output_photo_red = np.zeros((width, height, 1))
    output_photo_green = np.zeros((width, height, 1))
    output_photo_blue = np.zeros((width, height, 1))

    for j in range(0,width):
        for i in range(0,height):
            pixel = input_image[j,i]
            output_photo_blue[j,i] = pixel[0]
            output_photo_green[j,i] = pixel[1]
            output_photo_red[j,i] = pixel[2]


    return [output_photo_red, output_photo_green, output_photo_blue]
```

### Imagen Original

```python
images_dir = "./images/"
input_image = cv2.imread(images_dir + "input.jpg", cv2.IMREAD_COLOR)
rgb_input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
plt.axis("off")
plt.imshow(rgb_input_image)
plt.show()
```

### Funcin para separar canales

```python
output_file_red   =  images_dir + "salida_gray_rojo.jpg"
output_file_green =  images_dir + "salida_gray_verde.jpg"
output_file_blue  =  images_dir + "salida_gray_azul.jpg"

output_images = image_to_grayscales(input_image)
```

### Escala de grises canal rojo

```python
cv2.imwrite(output_file_red, output_images[0] )
output_image_red = cv2.imread(output_file_red, cv2.IMREAD_COLOR)
plt.imshow(output_image_red, cmap="gray")
```

### Escala de grises canal verde

```python
cv2.imwrite(output_file_green, output_images[1] )
output_image_green = cv2.imread(output_file_green,cv2.IMREAD_COLOR)
plt.imshow(output_image_green, cmap="gray")
```

### Escala de grises canal azul

```python
cv2.imwrite(output_file_blue, output_images[2] )
output_image_blue = cv2.imread(output_file_blue, cv2.COLOR_BGR2GRAY)
plt.imshow(output_image_blue, cmap="gray")
```
