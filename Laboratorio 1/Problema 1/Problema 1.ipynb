# Problema 1

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Se define la funci√≥n para aplicar el filtro deseado.

```python
def apply_filter(input_photo,color):
    width = input_photo.shape[0]
    height = input_photo.shape[1]
    output_photo = np.zeros((width, height, 3))
    red = 0
    green = 0
    blue = 0

    if color == 1 or color == 20 or color == 30:
        blue = 1
    
    if color == 2 or color == 10 or color == 20:
        green = 1
    
    if color == 3 or color == 10 or color == 30:
        red = 1
    
    for j in range(0,width):
        for i in range(0,height):
            pixel = input_photo[j,i]
            blue_pixel = blue * pixel[0]
            green_pixel = green * pixel[1]
            red_pixel = red * pixel[2]

            output_photo[j,i] = [blue_pixel,green_pixel,red_pixel]


    return output_photo
```

### Imagen Original

```python
input_image = cv2.imread("./images/saitama.jpg", cv2.IMREAD_COLOR)
rgb_input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_input_image)
```

### Filtro Azul

```python
output_file = "./images/output.jpg"

cv2.imwrite(output_file, apply_filter(input_image,1) )
output_image = cv2.imread(output_file, cv2.IMREAD_COLOR)
rgb_output_image = cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_output_image)
```
