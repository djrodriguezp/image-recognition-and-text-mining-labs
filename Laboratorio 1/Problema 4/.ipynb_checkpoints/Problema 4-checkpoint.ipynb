# Problema 4

Cree una funci√≥n que dada una imagen, muestre el histograma de cada canal de color y el de escala de grises.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def create_histograms(input_image):
    width = input_image.shape[0]
    height = input_image.shape[1]
    intensity = range(0,256)

    red_summatory = 0
    green_summatory = 0
    blue_summatory = 0

    frecuency_red = np.zeros((256))
    frecuency_green = np.zeros((256))
    frecuency_blue = np.zeros((256))

    for j in range(0,width):
        for i in range(0,height):
            pixel = input_image[j,i]
            frecuency_blue[pixel[0]]  += 1
            blue_summatory += pixel[0]

            frecuency_green[pixel[1]] += 1
            green_summatory += pixel[1]

            frecuency_red[pixel[2]]   += 1
            red_summatory += pixel[2]

    red_mean = red_summatory/(width * height)
    blue_mean = blue_summatory/(width * height)
    green_mean = green_summatory/(width * height)

    density_red = np.true_divide(frecuency_red, (width * height))
    density_green = np.true_divide(frecuency_green, (width * height))
    density_blue = np.true_divide(frecuency_blue, (width * height))

    plt.figure(figsize=[15,9])
    plt.subplot(222)
    plt.plot(intensity, density_red, 'r-')
    line1, = plt.plot([red_mean,red_mean], [0,max(density_red)], label="media: "+str(red_mean), linestyle='--', color='purple' )
    plt.ylabel("Densidad")
    plt.xlabel("Intensidad")
    plt.legend(handles=[line1])
    plt.title("Histograma Rojo")

    plt.subplot(221)
    plt.plot(intensity, density_green, 'g-')
    line2, = plt.plot([green_mean,green_mean], [0,max(density_green)], label="media: "+str(green_mean), linestyle='--',  color='purple')
    plt.ylabel("Densidad")
    plt.xlabel("Intensidad")
    plt.legend(handles=[line2])
    plt.title("Histograma Verde")
    
    plt.subplot(223)
    plt.plot(intensity, density_blue, 'b-')
    line3, = plt.plot([blue_mean,blue_mean], [0,max(density_blue)], label="media: "+str(blue_mean), linestyle='--', color='purple')
    plt.ylabel("Densidad")
    plt.xlabel("Intensidad")
    plt.legend(handles=[line3])
    plt.title("Histograma Azul")

    plt.savefig('histogramas.png')
    plt.show()
```

# Imagen de entrada

```python
input_image = cv2.imread("./images/input.jpg", cv2.IMREAD_COLOR)
rgb_input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_input_image)
```

```python
create_histograms(input_image)
```
