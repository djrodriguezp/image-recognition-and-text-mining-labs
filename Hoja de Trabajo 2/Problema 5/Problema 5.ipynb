# Problema 5
Desarrolle una funci贸n que dada una imagen a color y kernel de cualquier dimensi贸n (3X3, 5X5, 7X7, 9X9, 11X11) y cualquier tipo (gaussiano, Sobel, Sharp, unsharp, laplaciano), aplique un filtro de convoluci贸n a dicha imagen y muestre la imagen original y la imagen filtrada. Le recomendamos implementar el kernel como un arreglo de numpy.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Se define una funci贸n para aplicar un kernel a una imagen de entrada.

```python
def apply_filter(input_image, kernel):
    blue = 0
    green = 1
    red = 2

    height = input_image.shape[0]
    width  = input_image.shape[1]

    kernel_height = kernel.shape[0]
    kernel_width  = kernel.shape[1]
    kernel_center = kernel_width/2

    output_image =  np.zeros((height, width, 3))

    for j in range(0,height):
        for i in range(0, width):
            new_pixel = np.array([0, 0, 0])
            for m in range(0, kernel_height):
                for n in range(0, kernel_width):
                    row_offset = int(j + (m - kernel_center))
                    col_offset = int(i + (n - kernel_center))
                    if row_offset < 0 or row_offset >= height  :
                        continue
                    
                    if col_offset < 0 or col_offset >= width :
                        continue
                        
                    new_pixel[blue]  += (kernel[m, n] * int(input_image[row_offset, col_offset][blue]))
                    new_pixel[green] += (kernel[m, n] * int(input_image[row_offset, col_offset][green]))
                    new_pixel[red]   += (kernel[m, n] * int(input_image[row_offset, col_offset][red]))

            output_image[j,i][blue] = new_pixel[blue]
            output_image[j,i][green]  = new_pixel[green]
            output_image[j,i][red] = new_pixel[red]
            
    return output_image
```

### Imagen Original

```python
input_image = cv2.imread("./images/imagen.jpg", cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
plt.show()
```

### Gaussian blur 3x3

```python
kernel = np.array([[1, 2, 1], [2, 4, 2], [1, 2, 1]])
kernel = np.multiply(kernel, np.true_divide(1,16))

output_file = "./images/gaussian_blur_3x3.jpg"
cv2.imwrite(output_file, apply_filter(input_image, kernel) )
plt.imshow(cv2.cvtColor(cv2.imread(output_file, cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB))
plt.show()
```

### Gaussian blur 5x5

```python
kernel = np.array([[1, 4, 6, 4, 1], [4, 16, 24, 16, 4], [6, 24, 36, 24, 6], [4, 16, 24, 16, 4], [1, 4, 6, 4, 1]])
kernel = np.multiply(kernel, np.true_divide(1,256))

output_file = "./images/gaussian_blur_5x5.jpg"
cv2.imwrite(output_file, apply_filter(input_image, kernel) )
plt.imshow(cv2.cvtColor(cv2.imread(output_file, cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB))
plt.show()
```
