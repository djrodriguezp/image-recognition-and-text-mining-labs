# Problema 4
Implemente una función para cada una de las trasformaciones vistas en clase: negativa, lineal, logarítmica y exponencial.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def negative_transform(input_image):
    output_image = 255 - input_image
    return output_image
```

```python
def lineal_transform(input_image,alpha,beta):
    output_image = (alpha*input_image) + beta
    return output_image
```

```python
def log_transform(input_image):
    input_image = np.array(input_image, dtype = np.uint16)
    alpha = 255 / np.log(1 + np.max(input_image)) 
    output_image =  alpha * np.log(input_image)
    return np.array(output_image, dtype = np.uint8) 
```

```python
def exp_transform(input_image, alpha, gamma):
    output_image = alpha*(((input_image)/255)**gamma)
    return output_image*255
```

### Imagen Original

```python
input_image = cv2.imread("./images/imagen.jpg", cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
plt.show()
```

### Transformación Negativa

```python
output_file = "./images/negative.jpg"
cv2.imwrite(output_file, negative_transform(input_image) )
plt.imshow(cv2.cvtColor(cv2.imread(output_file, cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB))
plt.show()
```

### Transformación Lineal

```python
output_file = "./images/lineal_t.jpg"
cv2.imwrite(output_file, lineal_transform(input_image,0.2,100) )
plt.imshow(cv2.cvtColor(cv2.imread(output_file, cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB))
plt.show()
```

### Transformación Logarítmica

```python
output_file = "./images/logaritmica.jpg"
cv2.imwrite(output_file, log_transform(input_image) )
plt.imshow(cv2.cvtColor(cv2.imread(output_file, cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB))
plt.show()
```

```python
output_file = "./images/exponencial.jpg"
cv2.imwrite(output_file, exp_transform(input_image,0.2,3) )
plt.imshow(cv2.cvtColor(cv2.imread(output_file, cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB))
plt.show()
```
