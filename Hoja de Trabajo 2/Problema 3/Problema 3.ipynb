# Problema 3
Desarrolle una funci칩n que dada una imagen en escala de grises muestre dicha imagen en 3 dimensiones, donde los ejes X e Y representan el dominio espacial de la imagen y Z representa la intensidad del color gris de cada pixel.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

Se define una funci칩n para gr치ficar la intensidad de una imagen en un plano 3D.

```python
def plot_intensity(input_image, elev=None, azim=None):
    height = input_image.shape[0]
    width = input_image.shape[1]
    x = np.linspace(0, width, width)
    y = np.linspace(0, height, height)   
    xv, yv = np.meshgrid(x,y)
   
    fig = plt.figure(figsize=[10,8])
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(xv, yv, input_image, cmap='gray', linewidth=0)

    ax.view_init(elev,azim)
    ax.set_title("Gr치fica de intensidad en 3D")
    ax.set_xlabel('Ancho')
    ax.set_ylabel('Altura')
    ax.set_zlabel('Intensidad')
    ax.set_xlim([1,width])
    ax.set_ylim([1,height])
    ax.set_zlim([0,256])
    plt.show()
```

### Imagen Original

```python
input_image = cv2.imread("./images/imagen.jpg", cv2.IMREAD_COLOR)
input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)
plt.imshow(input_image, cmap='gray')
plt.show()
```

### Gr치fica de Intensidad en 3D

```python
plot_intensity(input_image)
```

```python
plot_intensity(input_image,90,90)
```
